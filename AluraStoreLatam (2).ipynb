#1. Análisis de facturación



```python
import pandas as pd

# URLs de tus archivos CSV
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar los DataFrames
try:
    tienda = pd.read_csv(url)
    tienda2 = pd.read_csv(url2)
    tienda3 = pd.read_csv(url3)
    tienda4 = pd.read_csv(url4)
except Exception as e:
    print(f"Error al cargar un archivo CSV: {e}. Asegúrate de que las URLs sean correctas.")
    exit() # Sale del script si hay un error de carga

# --- Cálculo de totales de facturación ---
total_facturacion_tienda_1 = tienda["Precio"].sum()
total_facturacion_tienda_2 = tienda2["Precio"].sum()
total_facturacion_tienda_3 = tienda3["Precio"].sum()
total_facturacion_tienda_4 = tienda4["Precio"].sum()

print(f"Total de facturación de Tienda 1: $ {round(total_facturacion_tienda_1,2)} ")
print(f"Total de facturación de Tienda 2: $ {round(total_facturacion_tienda_2,2)} ")
print(f"Total de facturación de Tienda 3: $ {round(total_facturacion_tienda_3,2)} ")
print(f"Total de facturación de Tienda 4: $ {round(total_facturacion_tienda_4,2)} ")


# Creando el diccionario de facturaciones

# Ahora, crearemos el diccionario `facturaciones` usando las variables que ya calculaste:


facturaciones = {
    "tienda1": total_facturacion_tienda_1,
    "tienda2": total_facturacion_tienda_2,
    "tienda3": total_facturacion_tienda_3,
    "tienda4": total_facturacion_tienda_4
}

# --- Encontrando la tienda con mayor y menor facturación ---
tienda_mayor_facturacion = max(facturaciones, key=facturaciones.get)
tienda_menor_facturacion = min(facturaciones, key=facturaciones.get)

print(f"\nLa tienda con mayor facturación fue: **{tienda_mayor_facturacion}** con ${facturaciones[tienda_mayor_facturacion]:,.2f}")
print(f"La tienda con menor facturación fue: **{tienda_menor_facturacion}** con ${facturaciones[tienda_menor_facturacion]:,.2f}")
```

# 3. Calificación promedio de la tienda


```python
def calificacion_total(tcf):

    calificacion_total = tcf['Calificación'].sum()
    return calificacion_total

def promedio_calificacion(pct):
    # Using a different variable name to avoid the conflict
    total_calificacion = calificacion_total(pct)
    promedio = total_calificacion / len(pct)
    return round(promedio, 2)


print('Total de puntos de tienda:', calificacion_total(tienda))
print('Promedio de Calificacion de tienda:', promedio_calificacion(tienda))
print('\n')
print('Total de puntos de tienda2:', calificacion_total(tienda2))
print('Promedio de Calificacion de tienda2:', promedio_calificacion(tienda2))
print('\n')
print('Total de puntos de tienda3:', calificacion_total(tienda3))
print('Promedio de Calificacion de tienda3:', promedio_calificacion(tienda3))
print('\n')
print('Total de puntos de tienda4:', calificacion_total(tienda4))
print('Promedio de Calificacion de tienda4:', promedio_calificacion(tienda4))
print('\n')
```

# 4. Productos más y menos vendidos

```python
import pandas as pd

# --- Datos de Facturación por Categoría en cada Tienda ---
# Convertimos los datos proporcionados en un diccionario de DataFrames
# Esto facilita el manejo y la aplicación de funciones por tienda

category_sales_data = {
    'Tienda 1': {
        'Electrónicos': 429493500.00, 'Electrodomésticos': 363685200.00,
        'Muebles': 187633700.00, 'Instrumentos musicales': 91299000.00,
        'Deportes y diversión': 39290000.00, 'Juguetes': 17995700.00,
        'Artículos para el hogar': 12698400.00, 'Libros': 8784900.00
    },
    'Tienda 2': {
        'Electrónicos': 410831100.00, 'Electrodomésticos': 348567800.00,
        'Muebles': 176426300.00, 'Instrumentos musicales': 104990300.00,
        'Deportes y diversión': 34744500.00, 'Juguetes': 15945400.00,
        'Artículos para el hogar': 14746900.00, 'Libros': 10091200.00
    },
    'Tienda 3': {
        'Electrónicos': 410775800.00, 'Electrodomésticos': 329237900.00,
        'Muebles': 201072100.00, 'Instrumentos musicales': 77380900.00,
        'Deportes y diversión': 35593100.00, 'Juguetes': 19401100.00,
        'Artículos para el hogar': 15060000.00, 'Libros': 9498700.00
    },
    'Tienda 4': {
        'Electrónicos': 409476100.00, 'Electrodomésticos': 283260200.00,
        'Muebles': 192528900.00, 'Instrumentos musicales': 75102400.00,
        'Deportes y diversión': 33350100.00, 'Juguetes': 20262200.00,
        'Artículos para el hogar': 15074500.00, 'Libros': 9321300.00
    }
}

# Convertir el diccionario en un DataFrame donde las filas son categorías y las columnas son tiendas
df_sales = pd.DataFrame(category_sales_data)

# --- Función para Identificar Productos Destacados ---
def get_most_least_sold_product(df_store_data):
    """
    Identifica la categoría de producto más y menos vendida para una tienda.

    Args:
        df_store_data (pd.Series): Una Serie de Pandas donde el índice es la categoría
                                   y los valores son las ventas.

    Returns:
        tuple: Una tupla que contiene:
               - (nombre_cat_mas_vendida, ventas_mas_vendida)
               - (nombre_cat_menos_vendida, ventas_menos_vendida)
    """
    if df_store_data.empty:
        return ("N/A", 0), ("N/A", 0)

    most_sold = df_store_data.idxmax(), df_store_data.max()
    least_sold = df_store_data.idxmin(), df_store_data.min()
    return most_sold, least_sold

# --- Almacenar y Mostrar Resultados del Análisis ---
# Diccionario para guardar los resultados del producto más y menos vendido por tienda
results = {}
for store_name, sales_series in df_sales.items():
    most_sold, least_sold = get_most_least_sold_product(sales_series)
    results[store_name] = {
        'Categoría Más Vendida': most_sold[0],
        'Ventas Más Vendidas ($)': most_sold[1],
        'Categoría Menos Vendida': least_sold[0],
        'Ventas Menos Vendidas ($)': least_sold[1]
    }

# Convertir los resultados a un DataFrame para una presentación tabular clara
df_results_analysis = pd.DataFrame.from_dict(results, orient='index')

print("--- **Análisis: Categorías de Productos Más y Menos Vendidas por Tienda** ---")
print(df_results_analysis.to_string(formatters={'Ventas Más Vendidas ($)': '$ {:,.0f}'.format,
                                                 'Ventas Menos Vendidas ($)': '$ {:,.0f}'.format}))
print("\n" + "="*80 + "\n")
```
